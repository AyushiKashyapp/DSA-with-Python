#### Quesion 1: o(n)
- First two statement of variable assignment takes constant time complexity because we are only assigning variables values.
- For 1st loop: Linear time complexity, traversing through each item of the element.
And tehn sum += i, o(1), since we are just assigning value to variable.
- Same for second loop.
- Printing values is also constant time operation.

```python
def foo(array):
    sum = 0 #o(1)
    product = 1 #o(1)

    for i in array: #o(n)
        sum += i #o(1)
    for i in array: #o(n)
        product *= i    #o(1)
    
    print(f"Sum : {sum}, Product: {product}") #o(1)
```

#### Question 2: o(n^2)

- Inner Loop: We have o(n) since traversing through n items of the array.
- Print function inside inner loop, has constant time complexity, o(1).
- Outer Loop makes the inner loop run n times, making the time complexity n^2.

```python
def printPairs(array):
    for i in array: #o(n^2)
        for j in array: #o(n)
            print(str(i) + "," + str(j)) #o(1)
```

#### Question 3: o(n^2)

**Way 1: Counting the iteration**

- 1st iteration: Will run n-1 times, where n is length of array.
- 2nd iteration: Will run n-2 times.
- 3rd iteration: Will run n-3 times.
- Will continue until we reach 1.

Summing up all these iteration to get the time complexity:
1 + 2 + ...+(n-3) + (n-2) + (n-1) = n(n-1)/2 = n^2 / 2 - n / 2 ----> We can remove the constant and the non-dominant term. Leaving us with n^2.

So it will be Quadratic Time Complexity, o(n^2).

**Way 2: Average Work**

- Outer loops runs n times.
- Inner loops runs different times for each outer loop.
    1st Outer loop run - Inner loop runs 10 times
    2nd Outer loop run - Inner loop runs 9 times
    Until we reach 1
    Average of all this will be 5 --> n/2

n * n/2 = n^2 / 2 -> o(n^2)


```python
def printUnorderedPairs(array):
    for i in range(0, len(array)): #o(n^2)
        for j in range(i + 1, len(array)):  #o(n)
            print(array[i] + "," + array[j]) #o(1)
```

#### Question 4: o(mn)

m = len(arrayA), n = len(arrayB)

Since we have nested loop, and both the loops run differently.

- First loop runs m times, and makes the inner loop also run. So, we have to take both complexities, o(mn).
- Inner loop run n times, which is length of the second array.
- The If condition and the print statement have constant time complexity.

```python
def printUnorderedPairs(arrayA, arrayB):
    for i in range(len(arrayA)): #o(m*n)
        for j in range(len(arrayB)): #o(n)
            if arrayA[i] < arrayB[j]: #o(1)
                print(str(arrayA[i]) + "," + str(arrayB[j])) #o(1)
```

#### Question 6: o(n)

- In the loop, we are going till the half of array so time complexity will be o(n/2), after removing the constant, o(n).
- In the rest of the steps, we are just assigning the values so o(1) for all those.

```python
def reverse(array):
    for i in range(0, int(len(array)/2)): #o(n/2) --> o(n)
        other = len(array) - i - 1 #o(1)
        temp = array[i] #o(1)
        array[i] = array[other] #o(1)
        array[other] = temp #o(1)
    print(array) #o(1)
```

#### Question 7: Which of teh following are equivalent to o(n)? why?

1. O(n + p), where p < n/2
2. O(2n)
3. O(n + log n)
4. O(n + n log n)
5. O(n + m)

#### Answers

1. o(n + p), where p < n/2: This means n > p, which means n is dominant and p is non dominant term, so can drop p -----> o(n).
2. o(2n): Drop constant ------> o(n)
3. o(n + log n): According to the Big O Complexity Chart, o(n) is dominant and o(log n) is non dominant, so we can drop it. -----> o(n)
4. o(n + n log n): According to the Big O Complexity Chart, o(n) is non dominant, so we can drop it ----> o(n log n)
5. o(n + m): Since, there is no established relationship between m and n, so we have check both variables. -----> Not Equivalent to o(n).
