Doubly Linked List

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def __iter__(self):
        node = self.head
        while node:
            yield node
            node = node.next

    def createLinkedList(self, value):
        node = Node(value)
        node.prev = None
        node.next = None
        self.head = node
        self.tail = node
        return "List Created!"
    
    def insertNode(self, location, value):
        if self.head is None:
            print("No node!")

        else:
            newNode = Node(value)
            if location == 0:
                newNode.next = self.head
                newNode.prev = None
                self.head.prev = newNode
                self.head = newNode

            elif location == -1:
                newNode.next = None
                newNode.prev = self.tail
                self.tail.next = newNode
                self.tail = newNode

            else:
                tempNode = self.head
                index = 0
                while index < location - 1:
                    tempNode = tempNode.next
                    index += 1
                newNode.next = tempNode.next
                newNode.prev = tempNode
                newNode.next.prev = newNode
                tempNode.next = newNode

    def traverseLinkedList(self):
        if self.head is None:
            print("No node!")

        else:
            temp = self.head
            while temp:
                print(temp)
                temp = temp.next

    def reverseTraverseLinkedList(self):
        if self.head is None:
            print("No node!")

        else:
            temp = self.head
            while temp:
                print(temp)
                temp = temp.prev

    def searchLinkedList(self, nodeValue):
        if self.head is None:
            print("No node!")

        else:
            temp = self.head
            while temp:
                if temp.value == nodeValue:
                    return True
                temp = temp.next
            return False
        
    def deleteNode(self, location):
        if self.head is None:
            print("No node!")

        else:
            if location == 0:
                if self.head == self.tail:
                    self.head = None
                    self.tail = None
                else:
                    self.head = self.head.next
                    self.head.prev = None
            elif location == -1:
                if self.head == self.tail:
                    self.head == None
                    self.tail == None
                else:
                    self.tail = self.tail.prev
                    self.tail.next = None
            else:
                current = self.head
                index = 0
                while index < current - 1:
                    current = current.next
                    index += 1
                current.next = current.next.next
                current.next.prev = current
            print("Deleted!")

    def deleteLinkedList(self):
        if self.head is None:
            print("No node!")

        else:
            current = self.head
            while current:
                current.prev = None
                current = current.next
            self.head = None
            self.tail = None
            print("All Deleted!")
```

```python
dll = LinkedList()
dll.createLinkedList(5)
dll.insertNode(0, 1)
dll.insertNode(1, 4)
dll.insertNode(2, 9)
print([node.value for node in dll])
```
